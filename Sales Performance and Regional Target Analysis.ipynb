# Sales Performance and Regional Target Analysis
A company wants to analyze monthly sales performance across different sales representatives and regions.
You are given three datasets:
1. sales_reps: sales representative information
2. monthly_sales: monthly sales records
3. regions: regional sales targets

Using these datasets, do the following:
- Inspect the datasets and identify their columns and dimensions.
- Identify the common key columns and merge the tables.
- Create a new column for total_sales by multiplying units sold by unit price.
- Compare each sales representative’s total sales with the regional target.
- Create a new column showing whether the target was met or not met.
- Calculate the total sales by region.
- Calculate the average total sales by sales representative.
- Create a pivot table showing total sales by month and region.
- Check for missing values, duplicated keys, and unmatched rows.
- Reshape the data if needed to improve the monthly comparison.

```python
import pandas as pd 

# Load the datasets
sales_reps = pd.read_excel("sales_reps.xlsx")

monthly_sales = pd.read_excel("monthly_sales.xlsx")

regions = pd.read_excel("regions.xlsx")

print("\nSales Representatives")
display(sales_reps)

print("\nMonthly Sales")
display(monthly_sales)

print("\nRegions")
display(regions)

# Identify their columns and dimensions
print("\nSales Representatives:", sales_reps.columns.tolist())
print("\nMonthly Sales:", monthly_sales.columns.tolist())
print("\nRegions:", regions.columns.tolist()) 

print("\nSales Representatives Shape", sales_reps.shape)
print("\nMonthly Sales Shape:", monthly_sales.shape)
print("\nRegions Shape:", regions.shape)


# Explain the common key used for merging
print("\nCommon key for merging (sales_reps vs montlhy_sales):", "rep_id")
print("\nCommon key for merging (sales_reps vs regions):", "region_id")
```

```python
import numpy as np

# Merge the sales_reps and monthly_sales tables
initial_sales_data = pd.merge(sales_reps, monthly_sales, on="rep_id", how="left")
sales_data = pd.merge(initial_sales_data, regions, on="region_id", how="left")

print("\nMerged Sales Data:")
display(sales_data)



# Create new column for total_sales
sales_data["total_sales"] = sales_data["units_sold"] * sales_data["unit_price"]

print("\nNew Total Sales Column")
display(sales_data[["rep_name", "month", "units_sold", "unit_price", "total_sales"]])



# Compare each sales reps total sales with regional target 
sales_data["target_met"] = np.where(sales_data["total_sales"] >= sales_data["monthly_target"], "Met", "Not Met")


print("\nSales Target Comparison Column")
display(sales_data[["rep_name", "total_sales", "region_name", "monthly_target", "target_met"]])
```

```python
# Calculate total sales by region 
sales_per_region = sales_data.groupby("region_name")["total_sales"].sum().sort_values(ascending=False)

print("\nTotal Sales per Region")
display(sales_per_region)


# Calculate average total sales by sales reps 
avg_sales_per_salesreps = sales_data.groupby("rep_name")["total_sales"].mean().sort_values(ascending=False)

print("\nAverage Total Sales per Sales Representatives")
display(avg_sales_per_salesreps)


# Create a pivot table that shows total sales by month and region 
pivot_table = pd.pivot_table(
    sales_data, 
    index="month",
    values="total_sales", 
    columns="region_name", 
    aggfunc="sum", 
    fill_value=0
)

print("\nPivot Table:")
display(pivot_table)
```

```python
# Check for missing values, duplicated keys, and unmatched rows. 

# First create a temporary merged tables using an outer join and add an indicator column 
first_sales_data = pd.merge(sales_reps, monthly_sales, on="rep_id", how="outer", indicator="merge_reps_sales")
temp_sales_data = pd.merge(first_sales_data, regions, on="region_id", how="outer", indicator="merge_with_regions")

print("\nTemporary Merged Sales Data:")
display(temp_sales_data) 

print("\nUnmatched Rows:")

print("\nRows Only in sales_reps:") 
display(temp_sales_data[temp_sales_data["merge_reps_sales"] == "left_only"]) 

print("\nRows Only in monthly_sales:") 
display(temp_sales_data[temp_sales_data["merge_reps_sales"] == "right_only"]) 

print("\nRows Only in final merge (regions):")
display(temp_sales_data[temp_sales_data["merge_with_regions"] != "both"])



# Check for missing values
missing_values = temp_sales_data.isna().sum()

print("\nMissing Values per column")
display(missing_values)


# Check for duplicated keys 
duplicated_keys = temp_sales_data.duplicated()

print("\nDuplicated Keys")
display(duplicated_keys)
```

```python
print("\nCurrent Sales Data")
display(sales_data)

# Reshape the main sales data to improe monthly comparison  

# Define month order 
month_order = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Convert month to ordered categorical
sales_data['new_month'] = pd.Categorical(sales_data['month'], categories=month_order, ordered=True)

# Create pivot table: months as rows, regions as columns, total sales as values
new_pivot_table = pd.pivot_table(
    sales_data,
    index='new_month',
    columns='region_name',
    values='total_sales',
    aggfunc='sum',
    observed=False
)

print("\nPivot Table for Monthly Comparison:")
display(new_pivot_table)
```
